```python
import csv
import os
import pandas as pd
```

```python
sub_list = []

for filename in os.listdir('./out'):
    if filename.endswith(".tsv"):
        sub  = filename[10:17]
        site = filename[6:9]
        sess = filename[18:20]
        rep  = filename[21:23]
        
        data = pd.read_csv(os.path.join('./out', filename), sep='\t')
        
        n_correct = len(data[data.accuracy == 1])
        n_wrong   = len(data[data.accuracy == 0])
        n_trials  = len(data.index)
        
        if n_trials != 0:
            percent_correct = n_correct/n_trials
        else:
            percent_correct = 0
            
        sub_list.append((sub, site, sess, rep, n_correct, n_wrong, n_trials, percent_correct))
    else:
        continue
        
df = pd.DataFrame(sub_list, columns=['id', 'site', 'session', 'repeat', 'correct', 'wrong', 'trials', 'percent_correct'])
duplicates = df[df.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicates)
```
